```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
ref_traj = pd.read_csv(filename1)
ref_x=ref_traj["x_coord"]
ref_y=ref_traj["y_coord"]
plt.plot(ref_x,ref_y)
#plt.plot(4,2,"o")
plt.legend(["Trajectory","Destination"])
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.savefig("trajectory.png",dpi=150)
```

```python
#Reading the PID CSV files
out=[]
with open(filename2, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    for row in csvreader:
        out.append(row)
#print(out)
print(float(out[2][1]))
print(out[1])
```

```python
#Super imposing trajectories
x=[float(out[1][i]) for i in range(0,len(out[1]))]
y=[float(out[2][j]) for j in range(0,len(out[1]))]
plt.plot(x,y)
plt.plot(ref_x,ref_y)
#plt.plot(4,2,"o",2,0.66,"s",3,1.22,"D")
#plt.plot(4,2,"o",2,0.66,"s",0.5,0.0,"D")
#plt.plot(4,2,"o",2,0.66,"s",1.33,0.22,"D")
plt.plot(4,2,"o",3,0.4,"D")
plt.legend(["Reference Trajectory","Traced Trajectory","Destination","Waypoint","Obstacle"])
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.savefig("trajectory_1_impose.png")
```
